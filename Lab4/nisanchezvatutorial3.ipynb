```python
class stock:
    def __init__(self, ticker, open, close, volume):
        self.ticker = ticker
        self.open = open
        self.close = close
        self.volume = volume
        self.rate_return = float(close)/open - 1
 
    def update(self, open, close):
        self.open = open
        self.close = close
        self.rate_return = float(self.close)/self.open - 1
 
    def print_return(self):
        print(self.rate_return)
```

```python
class stock2:
    def __init__(self, ticker, open, close, volume):
        self.ticker = ticker
        self.open = open
        self.close = close
        self.volume = volume
        self.rate_return = float(close)/open - 1
 
    def update(self, open, close):
        self.open = open
        self.close = close
        self.rate_return = float(self.close)/self.open - 1
 
    def print_return(self):
        print(self.rate_return)
```

```python
apple = stock('AAPL', 143.69, 144.09, 20109375)
google = stock('GOOG', 898.7, 911.7, 1561616)
```

```python
facebook = stock2('FB', 122.03, 123.09, 2010937)
twitter = stock2('TWTR', 76.2, 78.7, 1561616)
```

```python
apple.ticker
google.print_return()
google.update(912.8,913.4)
google.print_return()
```

```python
facebook.ticker
twitter.print_return()
twitter.update(77.1,77.4)
twitter.print_return()
```

```python
class child(stock):
    def __init__(self,name):
        self.name = name
```

```python
class child2(stock2):
    def __init__(self,name):
        self.name = name
```

```python
aa = child('aa')
print(aa.name)
aa.update(100,102)
print(aa.open)
print(aa.close)
aa.print_return()
```

```python
aa2 = child2('aa2')
print(aa2.name)
aa2.update(99,101)
print(aa2.open)
print(aa2.close)
aa2.print_return()
```
