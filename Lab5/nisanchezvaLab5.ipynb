#Para 2 A침os, con un presupuesto inicial de 100.000.000 y para la siguientes acciones : Facebook, Amazon, Google, Sony y Apple.
##Se eligieron estas acci칩nes debido al incremento de su precio en los ultimos a침os, obtuve esta informaci칩n a partir de busquedas en yahoo finance de distintas empresas. 

```python
class BootCampTask(QCAlgorithm):

    def Initialize(self):

        self.SetCash(100000000)
        self.SetStartDate(2019, 4, 20)
        self.SetEndDate(2021, 4, 20)

        
        self.fb = self.AddEquity("FB", Resolution.Minute)
        self.fb.SetDataNormalizationMode(DataNormalizationMode.Raw)
        self.amzn = self.AddEquity("AMZN", Resolution.Minute)
        self.amzn.SetDataNormalizationMode(DataNormalizationMode.Raw)
        self.goog = self.AddEquity("GOOG", Resolution.Minute)
        self.goog.SetDataNormalizationMode(DataNormalizationMode.Raw)
        self.sony = self.AddEquity("SONY", Resolution.Minute)
        self.sony.SetDataNormalizationMode(DataNormalizationMode.Raw)
        self.aapl = self.AddEquity("AAPL", Resolution.Minute)
        self.aapl.SetDataNormalizationMode(DataNormalizationMode.Raw)

        

    def OnData(self, data):
    
        
        if not self.Portfolio.Invested:
            self.SetHoldings("FB", 0.2)
            self.Debug(str(self.Portfolio["FB"].AveragePrice))
            self.SetHoldings("AMZN", 0.2)
            self.Debug(str(self.Portfolio["AMZN"].AveragePrice))
            self.SetHoldings("GOOG", 0.2)
            self.Debug(str(self.Portfolio["GOOG"].AveragePrice))
            self.SetHoldings("SONY", 0.2)
            self.Debug(str(self.Portfolio["SONY"].AveragePrice))
            self.SetHoldings("AAPL", 0.2)
            self.Debug(str(self.Portfolio["AAPL"].AveragePrice))
```
