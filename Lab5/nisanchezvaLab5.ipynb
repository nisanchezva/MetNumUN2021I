#Teniendo en cuenta el plazo de tiempo (2 años y 6 meses (10/10/2018 - 10/04/2021)) elegi las siguintes 5 acciones : 
##1.FaceBook : FB



```python
class BootCampTask(QCAlgorithm):

    def Initialize(self):
        #Dos años y 6 meses
        self.SetStartDate(2018, 10, 10)
        self.SetEndDate(2021, 4, 10)

        #1,2. Select IWM minute resolution data and set it to Raw normalization mode
        
        self.iwm = self.AddEquity("FB", Resolution.Minute)
        self.iwm.SetDataNormalizationMode(DataNormalizationMode.Raw)
        
    def OnData(self, data):

        #3. Place an order for 100 shares of IWM and print the average fill price
        #4. Debug the AveragePrice of IWM
        if not self.Portfolio.Invested:
            self.MarketOrder("FB", 100)
            self.Debug(str(self.Portfolio["FB"].AveragePrice))
```
